```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Cargar los datos
df_reviews = pd.read_csv('../data/interim/csv/reviews.csv')
```

```python
# Calculando el porcentaje de valores nulos por columna
missing_values_per_column = df_reviews.isnull().mean() * 100

print("\nHead\n")
display(df_reviews.head())
print("\nTail\n")
display(df_reviews.tail())

print("Info \n"); display(pd.DataFrame(df_reviews.info()))
print("\nTipos Unicos\n")
for column in df_reviews.columns:
    print(f'{column}: {df_reviews[column].apply(type).unique()}')

print("\nNulls \n", df_reviews.isnull().sum())

print("\nDescribe \n", df_reviews.describe())
display("\nPorcentaje de valores nulos por columna:\n", missing_values_per_column)
# Calcular la cantidad de valores nulos por fila
null_counts_per_row = df_reviews.isnull().sum(axis=1)

# Calcular la frecuencia de cada cantidad de valores nulos por fila
null_counts_frequencies = null_counts_per_row.value_counts(normalize=True).sort_index() * 100

# Crear un DataFrame para mostrar los resultados
null_counts_table = pd.DataFrame({
    'Cantidad de valores nulos por fila': null_counts_frequencies.index,
    'Porcentaje de filas': null_counts_frequencies.values
})
display(null_counts_table)
```

```python
# Obtener todos los valores nicos en la columna 'helpful'
valores_unicos = df_reviews['helpful'].unique()

# Utilizar un bucle para imprimir cada valor nico en una nueva lnea
for valor in valores_unicos:
    print(valor)
```
