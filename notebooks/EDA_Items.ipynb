```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datos
df_items = pd.read_csv('../data/interim/csv/items.csv')
```

```python
# Calculando el porcentaje de valores nulos por columna
missing_values_per_column = df_items.isnull().mean() * 100

print("\nHead\n")
display(df_items.head())
print("\nTail\n")
display(df_items.tail())

print("Info \n"); display(pd.DataFrame(df_items.info()))
print("\nTipos Unicos\n")
for column in df_items.columns:
    print(f'{column}: {df_items[column].apply(type).unique()}')

print("\nNulls \n", df_items.isnull().sum())

print("\nDescribe \n", df_items.describe())
display("\nPorcentaje de valores nulos por columna:\n", missing_values_per_column)
# Calcular la cantidad de valores nulos por fila
null_counts_per_row = df_items.isnull().sum(axis=1)

# Calcular la frecuencia de cada cantidad de valores nulos por fila
null_counts_frequencies = null_counts_per_row.value_counts(normalize=True).sort_index() * 100

# Crear un DataFrame para mostrar los resultados
null_counts_table = pd.DataFrame({
    'Cantidad de valores nulos por fila': null_counts_frequencies.index,
    'Porcentaje de filas': null_counts_frequencies.values
})
display(null_counts_table)
```
